```python
import pandas as pd
import numpy as np
```

```python
tdata = pd.read_csv("netpincer.csv")
```

```python
tdata.rename(index=str,columns={old : new for new, old in zip(['name','score','review_num'],tdata.columns)},inplace=True)
```

```python
tdata['score'] = tdata['score'].apply(lambda x: float(x))
tdata['review_num'] = np.where(tdata['review_num'].isna(), "0 értékelés", tdata['review_num'])
tdata['review_num'] = tdata['review_num'].apply(lambda x: int(x[:x.find(" ")]))
```

```python
df = tdata.sort_values(by='score',ascending=False).copy()
```

```python
highs = df[df.review_num > 200].sort_values(by='review_num',ascending=False)
```
